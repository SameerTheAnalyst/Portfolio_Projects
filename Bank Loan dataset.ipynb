# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading Dataset

```python
df = pd.read_csv('./loan_customer.csv')
df.head()
```

```python
nums = ['loan_id','birth_date','phone_number','dependents','applicant_income','coapplicant_income','loan_amount','loan_term_month','loan_term_year','credit_history']
cats = ['gender','married','education','self_employed','has_credit_card','property_type','loan_status']
```

# Data Cleaning

## Handling Missing Values

```python
#Let's look for any missing data

df.isna().sum()
```

```python
#Let's remove all the missing data

df = df.dropna()
```

# Handling Outlier

```python
print(f'Number of rows before filtering outliers: {len(df)}')

for col in ['applicant_income','coapplicant_income']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1
    low_limit = Q1 - 1.5*IQR
    high_limit = Q3 + 1.5*IQR
    df = df[(df[col] >= low_limit) & (df[col] <= high_limit)]

print(f'Number of rows after filtering outliers: {len(df)}')
```

# Feature Encoding

```python
df.head()
```

```python
#label encoding
mapping_gender = {
    'Female' : 0,
    'Male' : 1
}

mapping_married = {
    'No' : 0,
    'Yes' : 1
}

mapping_education = {
    'Not Graduate' : 0,
    'Graduate' : 1
}

mapping_self_employed = {
    'No' : 0,
    'Yes' : 1
}

mapping_has_credit_card = {
    'no' : 0,
    'yes' : 1
}

mapping_loan_status = {
    'REJECT' : 0,
    'APPROVE' : 1
}

df['gender'] = df['gender'].map(mapping_gender)
df['married'] = df['married'].map(mapping_married)
df['education'] = df['education'].map(mapping_education)
df['self_employed'] = df['self_employed'].map(mapping_self_employed)
df['has_credit_card'] = df['has_credit_card'].map(mapping_has_credit_card)
df['loan_status'] = df['loan_status'].map(mapping_loan_status)
```

```python
for cat in ['property_type']:
    onehots = pd.get_dummies(df[cat], prefix=cat)
    df = df.join(onehots)
```

```python
df = df.drop(columns=['property_type'])
```

```python
df.info()
```

# Business Insight

## Analyze customer interest based on property type

```python
df.groupby('gender')[['property_type_apartment', 'property_type_house', 'property_type_studio']].sum()
```

```python
#Analyzing customer property interests by gender through visualization


labels = ['Apartment','House','Studio']
men = [15,14,8]
women = [5,3,2]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men, width, label='Men')
rects2 = ax.bar(x + width/2, women, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Property Type')
ax.set_ylabel('Amount')
ax.set_title('Analyzing customer property interests by gender')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

Based on the data that has been analyzed, the results obtained are:

1. The number of applications for apartment property loans was 20 units, of which 15 units were from men and 5 units from women
2. The number of home property loan applications was 17 units, of which 14 units were from men and 3 units from women
3. The number of applications for studio property loans was 10 units, of which 8 units were from men and 2 units from women 

# Analysis of the influence of marital status on the borrowing period

```python
df.groupby(["loan_term_year", "married"])["loan_term_year"].count().to_frame()
```

```python
#Analysis of the influence of marital status on the borrowing period through visualization


labels = [7,15,25,30,40]
married = [1,3,0,23,0]
not_married = [0,0,1,18,1]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, married, width, label='Married')
rects2 = ax.bar(x + width/2, not_married, width, label='Not Married')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Borrowing Term (per year)')
ax.set_ylabel('Amount')
ax.set_title('Analysis of the influence of marital status on the borrowing period')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

Based on this analysis, insights have been obtained, namely:

1. In a loan with a term of 7 years, there is 1 married borrower
2. In a loan term of 15 years, there are 3 borrowers who are married
3. In a loan with a term of 25 years, there is 1 borrower who is not married
4. In the 30 year term loan, there are 41 borrowers, of which 23 people are married, and 18 people are not married.
5. In a loan with a term of 40 years, there is 1 borrower who is not married
